# Part 2: GroupBy Operations
## Q5. Aggregations:
- Group by type and count the number of titles in each category
- Group by country and find the top 5 countries with the most titles
- Group by rating and find average release year for each rating type

```python
type_counts = df.groupby('type')['title'].count()
print("Titles per type:\n", type_counts)
```

```python
top_countries = df.groupby('country')['title'].count().sort_values(ascending=False).head(5)

top_countries
```

```python
avg_release_by_rating = df.groupby('rating')['release_year'].mean().round(1)
print("\nAverage release year by rating:\n", avg_release_by_rating)
```

### Q6. Multiple aggregations:
- Group by release_year and show:
 - Total number of titles
 - Count of Movies and TV Shows separatel

```python
# Group by release_year with multiple aggregations
release_stats = df.groupby('release_year').agg(
    total_titles=('title', 'count'),
    movie_count=('type', lambda x: (x == 'Movie').sum()),
    tv_show_count=('type', lambda x: (x == 'TV Show').sum())
)

# Display the result
print(release_stats)
```

### Q7. Advanced:
- Create a new column `is_recent` True if `release_year` > 2015
- Group by country and count how many recent shows are there per country

```python
df['is_recent'] = df['release_year'] > 2015
recent_counts = df.groupby('country')['is_recent'].sum().sort_values(ascending=False)

print(df["is_recent"])
recent_counts.head()
```
