```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

img_array = np.array(data.astronaut())
plt.imshow(img_array)
plt.axis('off')
plt.title('Space')
plt.show()
```

### Crop

```python
croped_img = img_array[:300,:300,1]
plt.imshow(croped_img)
plt.axis('on')
plt.title('Space')
plt.show()
```

### Flip Image

```python
flipped_img = np.flip(img_array, axis=0)
plt.imshow(flipped_img,)
plt.axis('on')
plt.title('Space')
plt.show()
```

```python
rocket_img = np.array(data.rocket())
# gray_img = rocket_img.copy()

gray_img = 0.2989*rocket_img[:, :, 0] + 0.5870*rocket_img[:, :, 1] + 0.1140*rocket_img[:, :, 2]
plt.imshow(gray_img, cmap="gray")
plt.axis('off')
plt.title('Space')
plt.show()

print(rocket_img.shape)
print(rocket_img.ndim)
```
