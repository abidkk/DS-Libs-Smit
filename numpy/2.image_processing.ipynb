# Image Processing

```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
```

```python
astro = np.array(data.astronaut())
plt.imshow(astro)
plt.axis('off')
plt.title("Astronaut")
plt.show()
```

### Crop

```python
astrocopy = astro.copy()

croped_img = astrocopy[:300,:300] # rows, columns
plt.imshow(croped_img)
plt.axis('off')
plt.title('Astronaut Cropped')
plt.show()
```

```python
croped_img = astrocopy[:300,:200] # rows, columns 
plt.imshow(croped_img)
plt.axis('off')
plt.title('Astronaut Cropped')
plt.show()
```

### Flip Image

```python
flipped_img = np.flip(astrocopy, axis=0)
plt.imshow(flipped_img)
plt.axis('off')
plt.title('Astronaut Flipped')
plt.show()
```

# Colouring : RGB

```python
rocket_img = np.array(data.rocket())
coloured_img = rocket_img
# coloured_img[:, :,0] = 0 # red : 0 to 255
# coloured_img[:, :,1] = 250 # green : 0 to 255
coloured_img[:, :,2] = 100 # blue : 0 to 255

plt.imshow(coloured_img)
plt.axis('off')
plt.title('Coloured-1: RGB')
plt.show()
```

```python
coloured_img2 = rocket_img
coloured_img2[:, :,0] = 100 # red : 0 to 255
# coloured_img2[:, :,1] = 100 # green : 0 to 255
# coloured_img2[:, :,2] = 100 # blue : 0 to 255

plt.imshow(coloured_img2)
plt.axis('off')
plt.title('Coloured-2: RGB')
plt.show()
```

```python
coloured_img3 = rocket_img
coloured_img3[:, :,0] = 0 # red : 0 to 255
coloured_img3[:, :,1] = 0 # green : 0 to 255
coloured_img3[:, :,2] = 0 # blue : 0 to 255

plt.imshow(coloured_img3)
plt.axis('off')
plt.title('Coloured-3 black: RGB')
plt.show()
```

```python
coloured_img4 = rocket_img
coloured_img4[:, :,0] = 255 # red : 0 to 255
coloured_img4[:, :,1] = 255 # green : 0 to 255
coloured_img4[:, :,2] = 255 # blue : 0 to 255

plt.imshow(coloured_img4)
plt.axis('off')
plt.title('Coloured-4 white: RGB')
plt.show()
```

### RGB Gray

```python
coloured_img5 = rocket_img.copy()

gray_img = 0.2989*coloured_img5[:, :, 0] + 0.5870*coloured_img5[:, :, 1] + 0.1140*coloured_img5[:, :, 2]

plt.imshow(gray_img, cmap="gray")
plt.axis('off')
plt.title('Gray-5 : RGB')
plt.show()
```

```python
rocket_img = np.array(data.rocket())
gray_img = 0.2989*rocket_img[:, :, 0] + 0.5870*rocket_img[:, :, 1] + 0.1140*rocket_img[:, :, 2]
plt.imshow(gray_img, cmap="gray")
plt.axis('off')
plt.title('Space')
plt.show()
```
